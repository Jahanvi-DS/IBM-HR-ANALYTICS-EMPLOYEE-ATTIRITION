```python
# Import Important Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Data
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# Extract Important Columns for Analysis
print(df[['Age', 'Department', 'MonthlyIncome', 'Attrition']].head())
```

```python
# Data Mamipulation & Maths (Pandas + NumPy)
print("\n---  BUSINESS INSIGHTS LOGIC ---")

# Insight 1: Company overall Attrition Rate?
attrition_rate = df['Attrition'].value_counts(normalize=True) * 100
print(f"📊 Overall Attrition Rate: {attrition_rate['Yes']:.2f}% ")

# Insight 2: EMPLOYEE ATTRITION ACCORDING TO DEPARTMENT (Groupby)
dept_attrition = df.groupby('Department')['Attrition'].value_counts(normalize=True).unstack() * 100
print("\nDepartment wise Attrition %:")
print(dept_attrition)
```

```python
# 3. Data Visualization (Matplotlib + Seaborn)
plt.figure(figsize=(14, 5))

# Chart 1: Monthly Income vs Attrition (Seaborn Boxplot)
plt.subplot(1, 2, 1)
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, palette='Set1')
plt.title('Monthly Income vs Employee Attrition')
plt.xlabel('SKIP JOB? (Yes/No)')
plt.ylabel('Monthly Income ($)')

# Chart 2: Age vs Monthly Income (Seaborn Scatterplot + NumPy Trendline)
plt.subplot(1, 2, 2)
sns.scatterplot(x='Age', y='MonthlyIncome', hue='Attrition', data=df, alpha=0.7, palette='dark')
plt.title('Age aur Monthly Income ka Trend')
plt.xlabel('Age')
plt.ylabel('Monthly Income ($)')

plt.tight_layout()
plt.show()
```
